```python
import cPickle as pickle
import numpy as np
```

```python
with open("5000.pkl") as f:
    samples = pickle.load(f)
pos = np.array([[s.X, s.Y] for s in samples])
#the active samples at 5000 iterations
```

```python
import matplotlib as mpl
```

Plot of points

```python
mpl.pyplot.scatter(pos[:,0], pos[:,1])
mpl.pyplot.xlim(0, 200)
mpl.pyplot.ylim(0, 200)
mpl.pyplot.show()
```

8 component gmm

```python
from sklearn import mixture
import itertools
model = mixture.GMM(n_components=8)
model.fit(pos)
out = model.predict(pos)
color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm', 'y', 'k'])
for (i, color) in zip(range(len(model.means_)), color_iter):
    if not np.any(out == i):
        continue
    mpl.pyplot.scatter(pos[out == i, 0], pos[out == i, 1], .8, color=color)
```

DPGMM n_components = 10 means there are at most 10 components

```python
from sklearn import mixture
import itertools
model = mixture.DPGMM(n_components=10, alpha=1, covariance_type="full")
model.fit(pos)
out = model.predict(pos)
color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm', 'y', 'k'])
for (i, color) in zip(range(len(model.means_)), color_iter):
    if not np.any(out == i):
        continue
    mpl.pyplot.scatter(pos[out == i, 0], pos[out == i, 1], .8, color=color)
#print model.means_
```

VBGMM n_components = 15 means there are at most 15 clusters

```python
from sklearn import mixture
import itertools
model = mixture.VBGMM(n_components=15, alpha=0.01, covariance_type="full")
model.fit(pos)
out = model.predict(pos)
color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm', 'y', 'k'])
for (i, color) in zip(range(len(model.means_)), color_iter):
    if not np.any(out == i):
        continue
    mpl.pyplot.scatter(pos[out == i, 0], pos[out == i, 1], .8, color=color)
#print model.means_
```

K means with k = 10

```python
from sklearn import cluster
import itertools
model = cluster.KMeans(n_clusters=10)
model.fit(pos)
out = model.predict(pos)
color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm', 'y', 'k'])
for (i, color) in zip(range(len(model.cluster_centers_)), color_iter):
    if not np.any(out == i):
        continue
    mpl.pyplot.scatter(pos[out == i, 0], pos[out == i, 1], .8, color=color)
```

DBSCAN

```python
from sklearn import cluster
import itertools
model = cluster.DBSCAN()
out = model.fit_predict(pos)
color_iter = itertools.cycle(['r', 'g', 'b', 'c', 'm', 'y', 'k'])
for (i, color) in zip(range(len(model.components_)), color_iter):
    if not np.any(out == i):
        continue
    mpl.pyplot.scatter(pos[out == i, 0], pos[out == i, 1], .8, color=color)
```

Trying various possible GMMs with the number of components ranging from 1 to 20

```python
import itertools

import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
import matplotlib as mpl

X = pos

lowest_bic = np.infty
bic = []
n_components_range = range(1, 20)
cv_types = ['spherical', 'tied', 'diag', 'full']
for cv_type in cv_types:
    for n_components in n_components_range:
        # Fit a mixture of Gaussians with EM
        gmm = mixture.GMM(n_components=n_components, covariance_type=cv_type)
        gmm.fit(X)
        bic.append(gmm.bic(X))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm

bic = np.array(bic)
color_iter = itertools.cycle(['k', 'r', 'g', 'b', 'c', 'm', 'y'])
clf = best_gmm
bars = []

# Plot the BIC scores
spl = plt.subplot(2, 1, 1)
for i, (cv_type, color) in enumerate(zip(cv_types, color_iter)):
    xpos = np.array(n_components_range) + .2 * (i - 2)
    bars.append(plt.bar(xpos, bic[i * len(n_components_range):
                                  (i + 1) * len(n_components_range)],
                        width=.2, color=color))
plt.xticks(n_components_range)
plt.ylim([bic.min() * 1.01 - .01 * bic.max(), bic.max()])
plt.title('BIC score per model')
xpos = np.mod(bic.argmin(), len(n_components_range)) + .65 +\
    .2 * np.floor(bic.argmin() / len(n_components_range))
plt.text(xpos, bic.min() * 0.97 + .03 * bic.max(), '*', fontsize=14)
#spl.set_xlabel('Number of components')
spl.legend([b[0] for b in bars], cv_types)

# Plot the winner
splot = plt.subplot(2, 1, 2)
Y_ = clf.predict(X)
for i, (mean, covar, color) in enumerate(zip(clf.means_, clf.covars_,
                                             color_iter)):
    v, w = linalg.eigh(np.diag(covar))
    if not np.any(Y_ == i):
        continue
    plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], .8, color=color)

    # Plot an ellipse to show the Gaussian component
    angle = np.arctan2(w[0][1], w[0][0])
    angle = 180 * angle / np.pi  # convert to degrees
    v *= 4
    ell = mpl.patches.Ellipse(mean, v[0], v[1], 180 + angle, color=color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(.5)
    splot.add_artist(ell)

plt.xlim(0, 200)
plt.ylim(0, 200)
plt.xticks(())
plt.yticks(())
n_comp = len(clf.means_)
title = str(n_comp) + " components"
plt.title(title)
plt.subplots_adjust(hspace=.35, bottom=.2)
plt.show()
```
